# 3 columns conatin categorical data ie : Brand, Style and Country
we found that a lot of unique values within them...we need to reduce or drop those unimportant value so that reuce the space when we dumpify(转储) it.

```python
# get value counts of Brand
top_thirty_brands = df['Brand'].value_counts()[:30].index  # reduce to 30 by arbitrary (随意的), check & adjust later
top_thirty_brands
```

```python
# turn other brand to 'others'
# create a new column called newBrand
new_brand = []

for brand in df['Brand']:
    if brand in top_thirty_brands:
        new_brand.append(brand)
    else:
        new_brand.append('Other')
        
df['newBrand'] = new_brand
df
```

```python
# To dummyfy our data
brand_dummy = pd.get_dummies(df['newBrand'],drop_first=True,
                            prefix='from')
brand_dummy
```

```python
country_dummy = pd.get_dummies(df['newCountry'],drop_first=True,
                            prefix='in')
country_dummy
```
